```python
import pandas as pd

df = pd.read_csv("data/processed/airbnb_bookings_cleaned.csv")
```

Create features used in the pricing model

```python
monthly_demand_map = df.groupby("month")["month"].count().to_dict()

print(monthly_demand_map)
```

```python
import joblib

joblib.dump(monthly_demand_map, "src/monthly_demand_map.pkl")
```

```python
#Stay Category
def stay_category(n):
    if n == 1:
        return "single_night"
    elif n <= 3:
        return "short_stay"
    elif n <= 7:
        return "weekly_stay"
    else:
        return "long_stay"

df["stay_category"] = df["nights"].apply(stay_category)
```

```python
#Booking Window
def booking_window(days):
    if days <= 1:
        return "last_minute"
    elif days <= 7:
        return "short_notice"
    elif days <= 30:
        return "planned"
    else:
        return "early_booking"

df["booking_window"] = df["lead_time"].apply(booking_window)
```

```python
#Season Feature
def season(month):
    if month in [1,10]:
        return "high"
    elif month in [11,12]:
        return "holiday_peak"
    else:
        return "low_season"

df["season"] = df["month"].apply(season)
```

```python
#save dataset for ML models
df.to_csv("data/processed/airbnb_feature_engineered.csv", index=False)
```
