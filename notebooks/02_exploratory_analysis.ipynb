```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/processed/airbnb_bookings_cleaned.csv")
```

```python
#Bookings by months
monthly_bookings = df.groupby('month').size()

monthly_bookings.plot(kind='bar')

plt.title("Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.show()
```

Bookings peaks during the months of November to March

```python
# Average price per night for each listing
price_by_listing = df.groupby('listing')['price_per_night'].mean()

price_by_listing.plot(kind='bar')

plt.title("Average Price per Night by Listing")
plt.ylabel("Price")
plt.show()
```

```python
#Stay length distribution
df['nights'].hist(bins=20)

plt.title("Distribution of Stay Length")
plt.xlabel("Number of Nights")
plt.ylabel("Bookings")
plt.show()
```

Most guests stayed for 1-2 nights ,showing short leisure trips

```python
#LEAD TIME ANALYSIS
df['lead_time'].hist(bins=30)

plt.title("Booking Lead Time Distribution")
plt.xlabel("Days Before Check-in")
plt.show()
```

Most bookings occur 1 day before or same day as arrival, suggesting last-minute demand.


```python
#weekday arrivals
df['weekday'].value_counts().plot(kind='bar')
plt.title("Check-in Day Distribution")
plt.xlabel("Weekday")
plt.ylabel("Bookings")
plt.show()
```

Guests prefer weekend(Saturday>Friday>Sunday) check-ins.

```python
#Weeend or weekday departures
df['weekday_dept'].value_counts().plot(kind='bar')
plt.title("Check-out Day Distribution")
plt.xlabel("Weekday")
plt.ylabel("Bookings")
plt.show()
```

Guests prefer Sunday/Monday check-out

```python
# revenue by month
monthly_revenue = df.groupby('month')['gross_earnings'].sum()
monthly_revenue.plot(kind='bar')
plt.title("Revenue by Month")
plt.ylabel("Revenue")
plt.show()
```

Most Revenue came in during the months of November, December

```python
#Total revenue by listing
revenue_by_listing = df.groupby('listing')['gross_earnings'].sum()

revenue_by_listing.sort_values().plot(kind='barh')

plt.title("Total Revenue by Listing")
plt.xlabel("Revenue")
plt.show()
```

Most of the guests chose the listing for Room 6, Hence listings should be edited and made similar to Room 6

```python
#correlation between features
import seaborn as sns
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Feature Correlation")
plt.show()
```
